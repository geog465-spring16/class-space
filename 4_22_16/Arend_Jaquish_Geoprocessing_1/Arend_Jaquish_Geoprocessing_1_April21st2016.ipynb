```python
# Franz Arend and Spencer Jaquish
# GEOG 465
# April 21, 2016
# Geoprocessing 1: Basic geoprocessing with Python

# We have to import the system and arcpy stuff to create the layers, shapefiles, and do the operations.
import sys
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\bin')
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\arcpy')
sys.path.append('C:\\Program Files (x86)\\ArcGIS\\Desktop10.3\\ArcToolbox\\Scripts')

import arcpy
arcpy.env.workspace = "U:/GEOG 465/Arend_Jaquish_Geoprocessing_1/Seattle_schools.gdb"

#We used the search cursor and update cursor to select specifically the Seattle schools.
with arcpy.da.SearchCursor("Master_Table", "*") as cursor:
    for row in cursor:
            print row[1]
            
with arcpy.da.UpdateCursor("CBO_Geocode_1", "City") as cursor2:
    for row in cursor2:
            if row[0] != "Seattle":
                cursor2.deleteRow()

#We need to use the buffer analysis tool to create our buffer.
arcpy.Buffer_analysis("SE_schools", "U:/GEOG 465/Arend_Jaquish_Geoprocessing_1/Seattle_schools.gdb/SE_schools_buffered", "500 meters")

# First to make a select by location, we have to make a layer from the feature class
arcpy.MakeFeatureLayer_management("U:/GEOG 465/Arend_Jaquish_Geoprocessing_1/Seattle_schools.gdb/CBO_Geocode_1", "CBO_Geocode_1_lyr")

# Then we had to add the selection to the layer based on location to features in another feature class 
arcpy.SelectLayerByLocation_management ("CBO_Geocode_1_lyr", "within", "U:/GEOG 465/Arend_Jaquish_Geoprocessing_1/Seattle_schools.gdb/SE_schools_buffered")

#Lastly, we have to convert the feature class to another feature class which will allow us to make a shapefile out of it.
arcpy.FeatureClassToFeatureClass_conversion("CBO_Geocode_1_lyr", "U:/GEOG 465/Arend_Jaquish_Geoprocessing_1/Seattle_schools.gdb", "CBO_Geocode_Buffer")
```
