```python
import csv
```

```python
f = open(r'C:\Users\tal10\Downloads\parks_csv.csv') # Stores reference to Parks file
```

```python
reader = csv.reader(f, delimiter = ',') # Stores element to read the file
my_list = [] # Stores data from the Parks file
```

```python
# Adds data from Parks file to a list
next(reader, None)
for row in reader:
    my_list.append([row[0], row[1], row[2], row[3], row[4], row[5], 0])
```

```python
# Adds list with data to database to fill table
db.inserttable('parks', my_list)
```

```python
# Tests that data was added to database
print db.query('select * from parks where id < 11')
```

```python
f.close()
```
