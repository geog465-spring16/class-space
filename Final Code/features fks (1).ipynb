```python
db.query('alter table field_type drop constraint field_type_pkey')
```

```python
# Updates field_type id
db.query('UPDATE feature SET field_type_id = field_type.id \
         FROM field_type, seattle_parks_features \
         WHERE seattle_parks_features.feature_id = feature.id \
         AND seattle_parks_features.field_type = field_type.field_type \
         AND seattle_parks_features.id = feature.false_id')
```

```python
# Tests that the update was successful
print db.query("select * from feature")
```
