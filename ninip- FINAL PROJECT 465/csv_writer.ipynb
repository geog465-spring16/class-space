```python
import csv

open_input = open('C:/Users/ninip/Desktop/final_buildings.csv') #opens downloaded csv file

read_input = csv.DictReader(open_input) #reads the input file

project_info_output = open("C:/Users/ninip/Desktop/project_information.csv", 'w') #opens output file
permit_info_output = open("C:/Users/ninip/Desktop/permit_information.csv", 'w') #opens output file
application_info_output = open("C:/Users/ninip/Desktop/application_information.csv", 'w') #opens output file
location_output = open("C:/Users/ninip/Desktop/location.csv", 'w') #opens output file
building_output = open("C:/Users/ninip/Desktop/building_information_permits.csv", 'w') #opens output file

#creates empty lists
project_info = []
permit_info = []
application_info = []
location = []
building = []

#creates fieldnames
project_info_fieldnames = ['id','Application Permit', 'Action Type', 'Work Type', 'Value', 'Location_id']
permit_info_fieldnames = ['id','Application Permit','Permit Type', 'Description', 'Category', 'Contractor', 'Master Use Permit', 'Issue Date', 'Final Date', 'Expiration Date',\
                          'Location_id', 'Project_Information_id', 'Application_Information_id']
application_info_fieldnames = ['id', 'Application Permit', 'Applicant Name', 'Application Date', 'Status', 'Location_id', 'Project_Information_id']
location_fieldnames = ['id', 'Address', 'Latitude', 'Longitude']
building_fieldnames = ['id', 'Application Permit', 'Address', 'Location_id', 'Project_Information_id', 'Application_Information_id']

#populates lists
i = 1
for row in read_input:
    project_info.append({'id': i, 'Application Permit': row['Application/Permit Number'], 'Action Type': row['Action Type'], 'Work Type': row['Work Type'], 'Value': row['Value'], 'Location_id': 0})
    permit_info.append({'id': i, 'Application Permit': row['Application/Permit Number'], 'Permit Type': row['Permit Type'], 'Description': row['Description'], 'Category': row['Category'], 'Contractor': row['Contractor'],\
                        'Master Use Permit': row['Master Use Permit'], 'Issue Date': row['Issue Date'], 'Final Date': row['Final Date'], 'Expiration Date': row['Expiration Date'],\
                       'Location_id': 0, 'Project_Information_id': 0, 'Application_Information_id': 0})
    application_info.append({'id': i, 'Application Permit': row['Application/Permit Number'], 'Applicant Name': row['Applicant Name'], 'Application Date': row['Application Date'],\
                             'Status': row['Status'], 'Location_id': 0, 'Project_Information_id': 0})
    location.append({'id': i, 'Address': row['Address'], 'Latitude': row['Latitude'], 'Longitude': row['Longitude']})
    building.append({'id': i, 'Application Permit': row['Application/Permit Number'], 'Address': row['Address'], 'Location_id': i, 'Project_Information_id': i, 'Application_Information_id': i})
    i += 1
    
open_input.close()

project_info = sorted({project['Application Permit']:project for project in project_info}.values(), key=lambda k: k['Application Permit'])
permit_info = sorted({permit['Application Permit']:permit for permit in permit_info}.values(), key=lambda k: k['Application Permit'])
application_info = sorted({app['Application Permit']:app for app in application_info}.values(), key=lambda k: k['Application Permit'])
location = sorted({loc['Address']:loc for loc in location}.values(), key=lambda k: k['Address'])
building = sorted({build['Application Permit']:build for build in building}.values(), key=lambda k: k['Application Permit'])
    
#writes and populates csv files
project_dwriter = csv.DictWriter(project_info_output, fieldnames = project_info_fieldnames, delimiter = ",", lineterminator = "\n")
project_dwriter.writeheader()
project_dwriter.writerows(project_info) 

permit_dwriter = csv.DictWriter(permit_info_output, fieldnames = permit_info_fieldnames, delimiter = ",", lineterminator = "\n")
permit_dwriter.writeheader()
permit_dwriter.writerows(permit_info) 

application_dwriter = csv.DictWriter(application_info_output, fieldnames = application_info_fieldnames, delimiter = ",", lineterminator = "\n")
application_dwriter.writeheader()
application_dwriter.writerows(application_info) 

location_dwriter = csv.DictWriter(location_output, fieldnames = location_fieldnames, delimiter = ",", lineterminator = "\n")
location_dwriter.writeheader()
location_dwriter.writerows(location) 

building_dwriter = csv.DictWriter(building_output, fieldnames = building_fieldnames, delimiter = ",", lineterminator = "\n")
building_dwriter.writeheader()
building_dwriter.writerows(building)

#close files
project_info_output.close()
permit_info_output.close()
application_info_output.close()
location_output.close()
building_output.close()
```
